```python
#Employee Salary Increment
#A company gives a 5% yearly increment to an employee’s salary. Write a for loop to calculate the employee’s salary for the next 10 years starting
#from ₹50,000.

Salary=50000
print("Year \t Increment \t Salary")
for year in range(1,11):
    Increment=Salary+(Salary*0.05)
    Salary+=Increment
    print(year,"\t",round(increment, 2),"\t",round(Salary,2))
    
```

```python
#Class Attendance Percentage
#A student attended 18 out of 25 classes. Write a for loop to calculate and display the attendance percentage for each class attended
#(like running count).
Total_class=25
attended=18

print("Class\tAttendance %")

for i in range(1,attended+1):
    percentage = (i / Total_class) * 100
    print(i, "\t", round(percentage, 2))
```

```python
#Product Discount Prices
#A store applies a 10% discount on all products. Given a list of original prices [100, 250, 400, 560, 700], use a for loop to calculate and print the
#discounted prices

Prices=[100,250,400,560,700]
print("Price \t Discount_amount")

for price in Prices:
    discount = price * 0.10
    Final_amount=price-discount
    print(price, "\t", round(Final_amount, 2))
    
```

```python
#Monthly Savings Growth
#A person saves ₹1,000 every month. Write a for loop to calculate and display the total savings at the end of each month for 12 months

monthly_saving = 1000
total_saving = 0

print("Month\tTotal Savings")

for month in range(1, 13):
    total_saving += monthly_saving
    print(month, "\t", total_saving)
```

```python
#Electricity Bill Calculation
#The electricity cost is ₹5 per unit. If a household consumes [120, 135, 150, 160, 145] units over 5 months, use a for loop to calculate the bill for
#each month.

units_list = [120, 135, 150, 160, 145]
cost_per_unit = 5

print("Month\tUnits\tBill")

for i in range(len(units_list)):
    units = units_list[i]
    bill = units * cost_per_unit
    
    print(i+1, "\t", units, "\t", bill)
```

```python
#6. Student Marks Percentage
#A student scored marks in 5 subjects: [78, 85, 90, 66, 72]. Use a for loop to calculate the percentage (out of 100) for each subject.

marks = [78, 85, 90, 66, 72]
print("Subject\tMarks\tPercentage")

for i in range(len(marks)):
    percentage = (marks[i] / 100) * 100
    print(i+1, "\t", marks[i], "\t", percentage)
```
