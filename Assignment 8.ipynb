```python
#create all the 4 types of funtions for printing prime numbers 2 to 20
#without input and without returns
def prime():
    for num in range(2, 21):
        is_prime = True
        for i in range(2, num):
            if (num % i == 0):
                is_prime = False
                break
        if is_prime:
            print(num)

```

```python
prime()
```

```python
#with input and without return
def prime1(n1,n2):
    for num in range(n1,n2):
        prime=True
        for j in range(2,num):
            if (num%j==0):
                prime=False
                break
        if prime:
            print(num)
```

```python
prime1(2,21)
```

```python
#without input and with return
def prime3():
    for i in range(2,21):
        prime=True
        for j in range(2,i):
            if(i%j==0):
                prime=False
                break
        if prime:
            print(i)
    return prime3
    
```

```python
prime3()
```

```python
#with input and with return

def prime4(st, end):
    primes = []
    for i in range(st, end + 1):
        if (i < 2):
            continue
        prime = True
        for j in range(2, i):
            if (i % j == 0):
                prime = False
                break
        if prime:
            primes.append(i)
    return primes

result = prime4(2, 20)
print(result)
```
